```python
# importing libraries
import pandas as pd
import numpy as np
import glob
from datetime import datetime
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#importing all csv files
AllCSV = glob.glob('*.csv')
AllCSV
```

# Data Cleaning

```python
# reading all dataframes
all_dataframes = []
for csv in AllCSV:
    df = pd.read_csv(csv, encoding="latin1")
    df['country'] = csv[0:2]
    all_dataframes.append(df)
all_dataframes[0].head()
```

```python
all_dataframes[0].dtypes
```

```python
#Fixing Data Types
for df in all_dataframes:
    # video_id 
    df['video_id'] = df['video_id'].astype('str') 
    
    #title
    df['title'] = df['title'].astype('str')
    
    #channel_title
    df['channel_title'] = df['channel_title'].astype('str')
    
    #category_id
    df['category_id'] = df['category_id'].astype(str) 
    
    #tags
    df['tags'] = df['tags'].astype('str')
    
    #thumbnail_link
    df['thumbnail_link'] = df['thumbnail_link'].astype('str') 
    
    #description
    df['description'] = df['description'].astype('str')
    
    #comments_disabled
    df['comments_disabled'] = df['comments_disabled'].astype('category') 

    #ratings_disabled
    df['ratings_disabled'] = df['ratings_disabled'].astype('category') 

    #video_error_or_removed
    df['video_error_or_removed'] = df['video_error_or_removed'].astype('category') 

    #trending date
    df['trending_date'] = pd.to_datetime(df['trending_date'].str.replace(".", "-"), format='%y-%d-%m')

    #publish_time
    df['publish_time'] = pd.to_datetime(df['publish_time'], errors='coerce', format='%Y-%m-%dT%H:%M:%S.%fZ')
```

```python
# expanding the public time column
for df in all_dataframes:
    df.insert(4, 'publish_date', df['publish_time'].dt.normalize())
    df['publish_time_only'] = df['publish_time'].dt.time
    df['publish_hour'] = df['publish_time'].dt.hour
    df['publish_weekday'] = df['publish_time'].dt.weekday
    df['publish_month'] = df['publish_time'].dt.month
    df['publish_year'] = df['publish_time'].dt.year
```

```python
#combining all dataframes into a single big dataframe
combined_df = pd.concat(all_dataframes)
```

```python
# Making copy of original dataframe
backup_df = combined_df.reset_index().sort_values('trending_date', ascending=False).set_index('video_id')
backup_df.head()
```

```python
# Sorting according to latest trending date while removing duplicates
combined_df = combined_df.reset_index().sort_values('trending_date', ascending=False).drop_duplicates('video_id', keep='first').set_index('video_id')
combined_df.head()
```

```python
# sorting and resetting index for dataframes list
for df in all_dataframes:
    df = df.reset_index().sort_values('trending_date', ascending=False).set_index('video_id')
all_dataframes[2].head()
```

### General Overview

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Top categories
category_counts = (
    combined_df.groupby('category')
    .size()
    .reset_index(name='count')
)
category_counts = category_counts.sort_values('count', ascending=False).reset_index(drop=True)
category_counts
```

```python
# Bar chart visualizing top categories
plt.figure(figsize=(12,6))
sns.barplot(data=category_counts, x='count', y='category')
plt.xticks(rotation=45, ha='right')
plt.title=("Number of Videos per Category")
plt.xlabel=("count")
plt.ylabel=("category")
plt.tight_layout()
plt.show()
```

```python
# Top channels
combined_df['channel_title'].value_counts().head(10)
```

```python
# Videos per country
country_count = combined_df.groupby('country').agg(count=('country', 'count')).sort_values('count', ascending=False).reset_index()
country_count
```

```python
# Bar chart visualizing top countries
plt.figure(figsize=(12,6))
sns.barplot(data=country_count, x='count', y='country')
plt.xticks(rotation=45, ha='right')
plt.title = ("Number of Videos per Category")
plt.xlabel = ("count")
plt.ylabel = ("country")
plt.tight_layout()
plt.show()
```

### Engagement analysis

```python
# likes per category
likes_count = combined_df.groupby('category').agg(likes_count=('likes', 'sum')).sort_values('likes_count', ascending=False).reset_index()
#likes_count
```

```python
#dislikes per category
dislikes_count = combined_df.groupby('category').agg(dislikes_count=('dislikes', 'sum')).sort_values('dislikes_count', ascending=False).reset_index()
#dislikes_count
```

```python
#views per category
views_count = combined_df.groupby('category').agg(views_count=('views', 'sum')).sort_values('views_count', ascending=False).reset_index()
#views_count
```

```python
#comments per category
#views per category
comment_count = combined_df.groupby('category').agg(comment_count=('comment_count', 'sum')).sort_values('comment_count', ascending=False).reset_index()
#comment_count
```

```python
# Ratio of likes-dislikes in different categories
combined_df['like_ratio'] = (combined_df['likes'] / combined_df['dislikes'].replace(0, 1))
ratio_by_category = (
    combined_df.groupby('category')['like_ratio']
    .mean()
    .reset_index()
    .sort_values('like_ratio', ascending=False)
)
#ratio_by_category
```

```python
plt.figure(figsize=(12,6))
ax = sns.barplot(
    data=ratio_by_category,
    x= 'like_ratio',
    y='category',
    hue = 'category',
    palette = 'viridis',
    dodge = False,
    legend = False
)

plt.tight_layout()
plt.show()
    
```

```python
# Correlation between views, likes and comments
combined_df[['views', 'likes', 'comment_count']].corr()
```
